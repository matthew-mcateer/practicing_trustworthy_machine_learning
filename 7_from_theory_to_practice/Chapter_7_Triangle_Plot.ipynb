```python
#from IPython.display import set_matplotlib_formats
#set_matplotlib_formats('svg')
%config InlineBackend.figure_format ='retina'

import numpy as np
import matplotlib.pyplot as plt
import ternary

np.random.seed(1) #fix seed for notebook
```

### Functions for Ternary plotting

```python
plt.rcParams['figure.figsize'] = [15, 15]

## Boundary and Gridlines
scale = 40 #@param
figure, tax = ternary.figure(scale=scale)

# Draw Boundary and Gridlines
#tax.boundary(linewidth=2.0)
tax.gridlines(color="black", multiple=5)
tax.gridlines(color="blue", multiple=1, linewidth=0.5)

# Set Axis labels and Title
fontsize = 20 #@param
tax.set_title("Simplex Boundary and Gridlines", fontsize=fontsize)
tax.left_axis_label("Left label $\\alpha^2$", fontsize=fontsize)
tax.right_axis_label("Right label $\\beta^2$", fontsize=fontsize)
tax.bottom_axis_label("Bottom label $\\Gamma - \\Omega$", fontsize=fontsize)

# Set ticks
tax.ticks(axis='lbr', linewidth=1)

# Remove default Matplotlib Axes
tax.clear_matplotlib_ticks()

ternary.plt.show();
```

### The three metrics

```python
import math
import ternary

def shannon_entropy(p):
    """Computes the Shannon Entropy at a distribution in the simplex.
    Args:
      p (list): List of length 3 containing details of each axis
    """
    s = 0.
    for i in range(len(p)):
        try:
            s += p[i] * math.log(p[i])
        except ValueError:
            continue
    return -1.*s

def uncertainty(p):
    """Computes the Uncertainty at a distribution in the simplex.
    Args:
      p (list): List of length 3 containing details of each axis
    """
    return 1 - np.maximum.reduce([p[0], p[1], p[2]])

def class_margin(p):
    """Computes the classification margin, the probability of the second-most 
    likely class, at a distribution in the simplex.
    Args:
      p (list): List of length 3 containing details of each axis
    """
    p = np.array(p)
    part = np.partition(-p, 1)
    margin = - part[0] + part[1]
    return -margin + 1.0
```

### Plotting our 3 measures together

```python
fig = ternary.plt.figure(figsize=(30, 10))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

fontsize = 20  # @param

tax1 = ternary.TernaryAxesSubplot(ax=ax1, scale=60)
tax1.heatmapf(
    shannon_entropy, boundary=True, style="triangular", cmap="rainbow"
)
tax1.boundary(linewidth=1.0)
tax1.set_title("Shannon Entropy Heatmap", fontsize=fontsize)
tax1.ax.axis("equal")
tax1.ax.axis("off")

tax2 = ternary.TernaryAxesSubplot(ax=ax2, scale=60)
tax2.heatmapf(uncertainty, boundary=True, style="triangular", cmap="rainbow")
tax2.boundary(linewidth=1.0)
tax2.set_title("Uncertainty Heatmap", fontsize=fontsize)
tax2.ax.axis("equal")
tax2.ax.axis("off")

tax3 = ternary.TernaryAxesSubplot(ax=ax3, scale=60)
tax3.heatmapf(class_margin, boundary=True, style="triangular", cmap="rainbow")
tax3.boundary(linewidth=1.0)
tax3.set_title("Marginal Probability Heatmap", fontsize=fontsize)
tax3.ax.axis("equal")
tax3.ax.axis("off")

ternary.plt.show()
```
