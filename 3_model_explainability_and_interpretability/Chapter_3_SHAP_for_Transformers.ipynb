## 📜 Text Classification

```python
import shap
import transformers

from pprint import pprint
from transformers import (AutoTokenizer, 
                          AutoModelForSequenceClassification, 
                          TextClassificationPipeline)


tokenizer_name = "distilbert-base-uncased"

# This model is a fine-tuned version of distilbert-base-uncased 
# on the sst-2-english dataset.
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Switching out this model name, even for other distilbert models finetuned on
# sst-2 english, can result in changes to the visualization and output labels

tokenizer = transformers.DistilBertTokenizerFast.from_pretrained(tokenizer_name)
model = transformers.DistilBertForSequenceClassification.from_pretrained(
    model_name
).cpu()

pipe = TextClassificationPipeline(
    model=model, tokenizer=tokenizer, return_all_scores=True
)

def score_and_visualize(text):
    prediction = pipe([text])
    pprint(prediction[0])

    explainer = shap.Explainer(pipe)
    shap_values = explainer([text])

    shap.plots.text(shap_values)
```

```python
score_and_visualize(
    'After tons of trial and error, I finally succeeded in opening '
    'a shawarma place run by machine learning. The road was long '
    'and tough, but I am so happy with the outcome!'
)
```

```python
score_and_visualize('I am neutral to the restaurant.')
```

```python
score_and_visualize('I am impartial to the restaurant.')
```

```python
restaurant_review = (
    "This is easily the most underrated eatery this side of Cambridge, "
    "though that is also setting a low bar. The food was pretty good, "
    "but I did not like the service. The wait staff were really slow "
    "and did not seem to know what they were doing. The restaurant "
    "was really dirty and the owners did not seem to care. Still, I "
    "loved the food. It was amazing. As a flipside of the owners not "
    "caring, the establishment seemed remarkably dog-friendly. Some "
    "people will love this place. Some people will hate it. As for "
    "a final review, I won't not not give this place a good review."
)

score_and_visualize(restaurant_review)
```

## 🚀 Extending to Zero-shot classification

```python
import shap
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    ZeroShotClassificationPipeline,
)
from typing import Union, List

# distilbart-mnli is the distilled version of bart-large-mnli created using the 
# No Teacher Distillation technique proposed for BART summarisation by Huggingface
weights = "valhalla/distilbart-mnli-12-9"

model = AutoModelForSequenceClassification.from_pretrained(weights)
tokenizer = AutoTokenizer.from_pretrained(weights)

# Create your own pipeline that only requires the text parameter
# for the __call__ method and provides a method to set the labels

class MyZeroShotClassificationPipeline(ZeroShotClassificationPipeline):
    # Overwrite the __call__ method
    def __call__(self, *args):
        o = super().__call__(args[0], self.workaround_labels)[0]

        return [
            [
                {"label": x[0], "score": x[1]}
                for x in zip(o["labels"], o["scores"])
            ]
        ]

    def set_labels_workaround(self, labels: Union[str, List[str]]):
        self.workaround_labels = labels


labels = ["weather", "sports", "entertainment", "markets", "politics"]

model.config.label2id.update({v: k for k, v in enumerate(labels)})
model.config.id2label.update({k: v for k, v in enumerate(labels)})

pipe = MyZeroShotClassificationPipeline(
    model=model, tokenizer=tokenizer, return_all_scores=True
)
pipe.set_labels_workaround(labels)

def score_and_visualize(text):
    prediction = pipe([text])
    pprint(prediction[0])
    explainer = shap.Explainer(pipe)
    shap_values = explainer([text])
    shap.plots.text(shap_values)


example_text = (
    f"This is an example text about snowflakes in the summer "
    f"before election season and after football season."
)
score_and_visualize(example_text)
```

```python
example_text = "This is an example text about nothing at all."
score_and_visualize(example_text)
```

## 🛑 Downsides of SHAP


From these demonstrations, even for an esoteric task like zero-shot classification, SHAP seems really capable.

However, it should be noted that there are a few downsides to SHAP that you should consider first:

### Speed

SHAP is unfortunately a very slow interpretability algorithm.
One of the reasons we demonstrated this on text was that many implementatons of SHAP do not scale up from a 1-D array of tokens to a 2-D array of pixels (at least for anything larger than MNIST).

There are a few projects, such as [Fast TreeSHAP](https://github.com/linkedin/fasttreeshap), that have tried to speed up SHAP, but there are still other downsides that might give you pause.

### Mis-attribution

SHAP assigns equal weights to feature coalitions of all sizes.
But larger coalitions are often more informative.
Clever weighting by coalition size can be much better for finding valuable features. 
This was one of the motivations behind a SHAP variant called [WeightedSAP](https://github.com/ykwon0407/WeightedSHAP)
Still, this isn't a perfect solution, and SHAP can still be fooled and give false impressions of confidence.
