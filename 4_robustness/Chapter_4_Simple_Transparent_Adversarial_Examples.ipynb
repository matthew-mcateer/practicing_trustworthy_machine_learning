```python
%matplotlib inline

import os

import numpy as np
import tensorflow as tf
tf.compat.v1.disable_eager_execution()

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50
from art.estimators.classification import KerasClassifier
```

We then load the classifier.

```python
#@title model
mean_imagenet = np.zeros([224, 224, 3])
mean_imagenet[...,0].fill(103.939)
mean_imagenet[...,1].fill(116.779)
mean_imagenet[...,2].fill(123.68)
model = ResNet50(weights='imagenet')
classifier = KerasClassifier(
    clip_values=(0, 255), 
    model=model, 
    preprocessing=(mean_imagenet, np.ones([224, 224, 3]))
)
```

We load the data now. Make sure to install [Imagenet Stubs](https://github.com/nottombrown/imagenet-stubs) before proceeding.

```python
# @title data
import pandas as pd
from imagenet_stubs import get_image_paths
from imagenet_stubs.imagenet_2012_labels import label_to_name, name_to_label

all_paths = get_image_paths()
all_imgs = []
for path in all_paths:
    img = image.load_img(path, target_size=(224, 224))
    img = image.img_to_array(img)
    all_imgs.append(img)
all_names = [os.path.basename(path) for path in all_paths]
all_probs = [np.max(classifier.predict(np.array([img]))) for img in all_imgs]
all_labels = [
    np.argmax(classifier.predict(np.array([img]))) for img in all_imgs
]

img_data = pd.DataFrame(
    {
        "name": [os.path.splitext(name)[0] for name in all_names],
        "label": [label_to_name(label) for label in all_labels],
        "prob": [round(p, 3) for p in all_probs],
        "img": all_imgs,
    }
).set_index("name")
```

Before proceeding, let us check the majority class probabilities for each image predicted by our classifier.

```python
img_data[['label','prob']].sort_values(['prob']).reset_index()
```

We define `SimpleTransparentExamples` as our attack class, then apply the attack through setting 1 and 2.

```python
#@title simple transparent adversarial examples
from PIL import Image, ImageDraw, ImageFont

!wget -q -O sans_serif.ttf https://github.com/matthew-mcateer/practicing_trustworthy_machine_learning/raw/main/4_robustness/sans_serif.ttf

class SimpleTransparentExamples:
    def __init__(
        self, image, text, font=ImageFont.truetype("sans_serif.ttf", 16)
    ):
        self.image = image
        self.text = text
        self.font = font

    def generate(
        self, alpha=0.3, angle=0, x=0, y=0, image_init=None, resume=False
    ):
        # watermark
        opacity = int(256 * alpha)
        mark_width, mark_height = self.font.getsize(self.text)
        patch = Image.new("RGBA", (mark_width, mark_height), (0, 0, 0, 0))
        ImageDraw.Draw(patch).text(
            (0, 0), text=self.text, font=self.font, fill=(255, 0, 0, opacity)
        )
        patch = patch.rotate(angle, expand=1)

        # merge
        wx, wy = patch.size
        if resume == True:
            img = image.array_to_img(image_init)
        else:
            img = self.image.copy()
        img.paste(patch, (x, y, x + wx, y + wy), patch)
        img = image.img_to_array(img)
        return img
```

```python
#@title setting 1: alpha=.1 (NO repeat text)
nm = "beagle" #@param
font_size_1 = 16 #@param
font = ImageFont.truetype("sans_serif.ttf", font_size_1)
patch_text = "Hello World" #@param
alpha = 0.1 #@param
angle = 30 #@param

# load images
all_raw = [image.load_img(path, target_size=(224, 224)) for path in all_paths]
raw_data = pd.DataFrame(
    {"name": [os.path.splitext(name)[0] for name in all_names], "img": all_raw}
).set_index("name")

im = raw_data.loc[nm, "img"].copy()

si = SimpleTransparentExamples(image=im, text=patch_text, font=font)
xlist = []
ylist = []
imlist = []
for x in np.arange(0, 224, 20):
    for y in np.arange(0, 224, 20):
        img = si.generate(alpha=alpha, angle=angle, x=x, y=y, resume=False)
        xlist.append(x)
        ylist.append(y)
        imlist.append(img)

        print(
            [
                x,
                y,
                label_to_name(np.argmax(classifier.predict(np.array([img])))),
            ]
        )

outputs_s1 = pd.DataFrame({"x": xlist, "y": ylist, "img": imlist})
```

```python
#@title setting 2: alpha=.05 (repeat text)
font_size_2 = 8 #@param
font = ImageFont.truetype("sans_serif.ttf", font_size_2)
alpha = 0.05 #@param
im = raw_data.loc[nm, "img"].copy()

si = SimpleTransparentExamples(image=im, text=patch_text, font=font)
xlist = []
ylist = []
imlist = []
for x in np.arange(0, 224, 20):
    for y in np.arange(0, 224, 20):
        img = si.generate(
            alpha=alpha, angle=angle, x=x, y=y, resume=True, image_init=img
        )
        xlist.append(x)
        ylist.append(y)
        imlist.append(img)

        print(
            [
                x,
                y,
                label_to_name(np.argmax(classifier.predict(np.array([img])))),
            ]
        )

outputs_s2 = pd.DataFrame({"x": xlist, "y": ylist, "img": imlist})
```

```python
from textwrap import wrap
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 15), dpi=100)

plt.subplot(1, 3, 1)
plt.imshow(img_data.loc[nm, "img"].astype(np.uint))
plt.title("Original")
orig_label = img_data.loc[nm, "label"]
plt.xlabel("Label: " + "\n".join(wrap(orig_label, 35)))  # wraps label

plt.subplot(1, 3, 2)
im1 = outputs_s1.loc[
    (outputs_s1["x"] == 40) & (outputs_s2["y"] == 20), "img"
].to_list()[0]
lab1 = label_to_name(np.argmax(classifier.predict(np.array([im1]))))
plt.imshow(im1.astype(np.uint))
plt.title("Setting 1, single text")
plt.xlabel("Label: " + "\n".join(wrap(lab1, 35)))

plt.subplot(1, 3, 3)
im2 = outputs_s2.loc[
    (outputs_s2["x"] == 200) & (outputs_s2["y"] == 20), "img"
].to_list()[0]
lab2 = label_to_name(np.argmax(classifier.predict(np.array([im2]))))
plt.imshow(im1.astype(np.uint))
plt.title("Setting 2, repeated text")
plt.xlabel("Label: " + "\n".join(wrap(lab2, 35)))

plt;
```
