```python
from transformers import (
    BertForSequenceClassification,
    BertTokenizerFast,
    DistilBertForSequenceClassification,
    Trainer,
    TrainingArguments,
)
from transformers import glue_compute_metrics as compute_metrics
from nlp import load_dataset
import torch
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

model_vic = BertForSequenceClassification.from_pretrained("bert-base-uncased")
model_copycat = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")


def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True)
```

The original work used the following datasets

- Blog posts dataset **(topic classification)** from the blog authorship corpus (Schler et al., 2006)
    - https://huggingface.co/datasets/blog_authorship_corpus
    - https://paperswithcode.com/dataset/blog-authorship-corpus
    - The Blog Authorship Corpus consists of the collected posts of 19,320 bloggers gathered from blogger.com in August 2004. The corpus incorporates a total of 681,288 posts and over 140 million words - or approximately 35 posts and 7250 words per person.
    - It is unfortunately not available, due to an error:
        - `NonMatchingChecksumError: Checksums didn't match for dataset source files: ['http://www.cs.biu.ac.il/~koppel/blogs/blogs.zip']`
- AG **(topic classification)** AG news corpus (Del Corso et al., 2005)
    - https://huggingface.co/datasets/ag_news
- TP-US **(sentiment analysis)** from Trustpilot Sentiment dataset (Hovy et al., 2015)
- Yelp **(sentiment analysis)** YELP dataset (Zhang et al., 2015)
    - https://huggingface.co/datasets/yelp_polarity
- IMDB (sentiment analysis)
    - https://huggingface.co/datasets/imdb

For our tutorial here, by default **we'll use the IMDB sentiment classification dataset**.

```python
#@title Setting up the IMDB training and test dataset splits
imdb_train_dataset, imdb_test_dataset = load_dataset(
    "imdb", split=["train", "test"]
)    
```

```python
#@title Tokenizing and formatting the training and test datasets
imdb_train_dataset = imdb_train_dataset.map(
    tokenize, batched=True, batch_size=len(imdb_train_dataset)
)
imdb_test_dataset = imdb_test_dataset.map(
    tokenize, batched=True, batch_size=len(imdb_test_dataset)
)
imdb_train_dataset.set_format(
    "torch", columns=["input_ids", "attention_mask", "label"]
)
imdb_test_dataset.set_format(
    "torch", columns=["input_ids", "attention_mask", "label"]
)
```

```python
#@title Computing metrics for the initial model
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average="binary"
    )
    acc = accuracy_score(labels, preds)
    return {
        "accuracy": acc,
        "f1": f1,
        "precision": precision,
        "recall": recall,
    }
```

We can then train the victim model.

It's worth noting that the success of the model theft will depend on the model hyperparameters. Some combinations of parameters may not be possible on some hardware. For reference, the initial work this tutorial is based on used the following hyperparameters to train the victim model:

- Underlying model: [base BERT (uncased)](https://huggingface.co/docs/transformers/model_doc/bert)
- task: the BLOG dataset (unavailable, see the above comment)
- lower-case normalization on input data
- Maximum sequence length of `128`
- per-GPU training batch size: `64`
- per-GPU test batch size: `64`
- Learning rate: `5e-5`
- Number of training epochs: `4`
- Seed value: `1234`

All of this with evaluating during training (which can be disabled to save time or compue).

```python
#@title Arguments for training the victim model

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    # evaluate_during_training=True,
    logging_dir="./logs",
)

trainer_vic = Trainer(
    model=model_vic,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=imdb_train_dataset,
    eval_dataset=imdb_test_dataset,
)
```

```python
#@title Training the victim model
#@markdown Can take about 40 minutes on a good GPU
trainer_vic.train()
```

```python
# Can take about 15 minutes to evaluate IDMB test dataset on a good GPU
trainer_vic.evaluate()
```

## Training the extracted model

```python
class ExtractDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {}
        item["attention_mask"] = torch.tensor(
            self.encodings[idx]["attention_mask"]
        )
        item["input_ids"] = torch.tensor(self.encodings[idx]["input_ids"])
        item["label"] = torch.tensor(
            self.labels[idx].argmax(-1), dtype=torch.long
        )
        return item

    def __len__(self):
        return len(self.labels)
```

```python
theft_train_dataset = ExtractDataset(
    sample_test_dataset, prediction_output.predictions
)     
```

```python
#@title Computing metrics for the copycat model
def compute_metrics_copycat(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average="binary"
    )
    acc = accuracy_score(labels, preds)
    return {
        "accuracy": acc,
        "f1": f1,
        "precision": precision,
        "recall": recall,
    }
```

We can then train the copycat model.

It's worth noting that the success of the model theft will depend on the hyperparameters. Some combinations of parameters may not be possible on some hardware. For reference, the initial work this tutorial is based on used the following hyperparameters to train the copycat model:

- Underlying model: [DistilBERT](https://huggingface.co/docs/transformers/model_doc/distilbert) ([BERT](https://huggingface.co/docs/transformers/model_doc/bert) was used as the victim model)
- task: The predictions produced by the initial victim model
- lower-case normalization on input data
- Maximum sequence length of `128`
- per-GPU training batch size: `64`
- per-GPU test batch size: `64`
- Learning rate: `5e-5`
- Number of training epochs: `4`
- Seed value: `1234`

All of this with evaluating during training (which can be disabled to save time or compue).

```python
#@title Arguments for training the theft model
theft_training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    # evaluate_during_training=True,
    logging_dir="./logs",
)

trainer_extract = Trainer(
    model=model_copycat,
    args=theft_training_args,
    compute_metrics=compute_metrics_copycat,
    train_dataset=theft_train_dataset,
    eval_dataset=imdb_test_dataset
)     
```

```python
#@title Training the copycat model
#@markdown Can take around 1 hour on a good GPU.
#@markdown Batch size needs to be increased beyond 16 for reliable reliable reduction in loss
trainer_extract.train()
```

```python
#@title Evaluating the copycat model
#@markdown takes about 15 minutes to run on IMDB test dataset
trainer_extract.evaluate()     
```

## How much does all this cost?

The original authors looked at the efficiency of MEA on various classification datasets with various inference service providers.
These inference service-providers charge according to a pay-as-you-use policy.

Below is a copy of Table 3 from the paper, which estimated costs according to [Google NLP APIs](https://cloud.google.com/natural-language/pricing) and [IBM Watson APIs](https://www.ibm.com/cloud/watson-natural-language-understanding/).

Table 3: Estimate costs of model extraction on different datasets

|                           **Dataset**                          | **#Query** | **Google price** | **IBM price** |
|:--------------------------------------------------------------:|:----------:|:----------------:|:-------------:|
| [TP-US]()                                                      | `22,142`   | `$22.10`         | `$66.30`      |
| [Yelp](https://huggingface.co/datasets/yelp_polarity)          | `~520K`    | `$520.00`        | `$1,560.00`   |
| [AG](https://huggingface.co/datasets/ag_news)                  | `~112K`    | `$112.00`        | `$336.00`     |
| [Blog](https://huggingface.co/datasets/blog_authorship_corpus) | `7,098`    | `$7.10`          | `$21.30`      |

If the target model in question is a BERT-based model, this approach to stealing the model is highly economical.

That being said, this analysis has NOT been performed on larger inference APIs such as those hosted by Cohere and/or OpenAI. 
Stealing a 175B-parameter model is a much more expensive task than stealing a model with parameters numbering in the millions.
This is especially the case when scaling up the size of the dataset to match the size of the new model.
